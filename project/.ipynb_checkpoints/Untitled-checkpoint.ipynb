```python
import geopandas as gpd
import pandas as pd
import os
import matplotlib.pyplot as plt

# Set the folder and file paths
folder = '/Users/maryamtorkashvand/Library/CloudStorage/OneDrive-UniversityofIowa/My_Phd/Phd_Courses/Spring_2024/Geoprogramming/project/data/'
shapefile = folder + 'states48hq.shp'
partitiondir = folder + 'flows/Leiden_partition'

# Load the shapefile
states48 = gpd.read_file(shapefile)
states48 = states48.to_crs("ESRI:102003")

# Load partition data
partitions = []
for dirpath, dirnames, files in os.walk(partitiondir):
    for file in files:
        if file.endswith('.csv'):
            file_path = os.path.join(dirpath, file)
            df = pd.read_csv(file_path)
            partitions.append(df)

# Function to calculate Jaccard Index
def calculate_jaccard(set1, set2):
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union if union else 0

# Initialize empty list to store geodataframes
maps = []
threshold = 0.2  # Jaccard index threshold for considering partitions as similar

# Iterate over partitions and shapefile to merge and process each
for i, partition in enumerate(partitions):
    # Merge with the shapefile on the 'rootsid' which matches 'node' in partitions
    gdf = states48.merge(partition.rename(columns={'node': 'rootsid', 'partition': f'partition_{i}'}), on='rootsid')
    
    if i > 0:
        # Calculate Jaccard index between current and previous partitions, map similar ones
        prev_gdf = maps[-1]
        current_labels = {}
        for label, group in gdf.groupby(f'partition_{i}'):
            best_match = None
            best_score = 0
            for prev_label, prev_group in prev_gdf.groupby(f'partition_{i-1}'):
                score = calculate_jaccard(set(group['rootsid']), set(prev_group['rootsid']))
                if score > best_score and score >= threshold:
                    best_match = prev_label
                    best_score = score
            if best_match:
                current_labels[label] = best_match
            else:
                current_labels[label] = f"Region {len(current_labels) + 1}"
        gdf[f'partition_{i}'] = gdf[f'partition_{i}'].map(current_labels)
    else:
        # Initialize partitions for the first map
        unique_partitions = gdf[f'partition_{i}'].unique()
        partition_to_region = {part: f"Region {j+1}" for j, part in enumerate(unique_partitions)}
        gdf[f'partition_{i}'] = gdf[f'partition_{i}'].map(partition_to_region)
    
    maps.append(gdf)

# Visualization of maps
fig, axes = plt.subplots(1, len(maps), figsize=(30, 10))

for i, gdf in enumerate(maps):
    gdf.plot(ax=axes[i], column=f'partition_{i}', legend=True)
    axes[i].set_title(f"Map {i + 1}")
    axes[i].set_axis_off()

plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import os
import matplotlib.pyplot as plt

# Set the folder and file paths
folder = '/Users/maryamtorkashvand/Library/CloudStorage/OneDrive-UniversityofIowa/My_Phd/Phd_Courses/Spring_2024/Geoprogramming/project/data/'
shapefile = folder + 'states48hq.shp'
partitiondir = folder + 'flows/Leiden_partition'

# Load the shapefile
states48 = gpd.read_file(shapefile)
states48 = states48.to_crs("ESRI:102003")

# Load partition data
partitions = []
for dirpath, dirnames, files in os.walk(partitiondir):
    for file in files:
        if file.endswith('.csv'):
            file_path = os.path.join(dirpath, file)
            df = pd.read_csv(file_path)
            partitions.append(df)

# Function to calculate Jaccard Index
def calculate_jaccard(set1, set2):
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union if union else 0

# Initialize empty list to store geodataframes
maps = []
new_region_start = 100  # Start new regions from a high number to avoid conflict with existing labels

# Iterate over partitions and shapefile to merge and process each
for i, partition in enumerate(partitions):
    # Merge with the shapefile on the 'rootsid' which matches 'node' in partitions
    gdf = states48.merge(partition.rename(columns={'node': 'rootsid', 'partition': f'partition_{i}'}), on='rootsid')
    
    if i > 0:
        # Calculate Jaccard index between current and previous partitions, map similar ones
        prev_gdf = maps[-1]
        current_labels = {}
        unmatched_labels = set(gdf[f'partition_{i}'].unique())  # Track unmatched labels

        for label, group in gdf.groupby(f'partition_{i}'):
            best_match = None
            best_score = 0
            for prev_label, prev_group in prev_gdf.groupby(f'partition_{i-1}'):
                score = calculate_jaccard(set(group['rootsid']), set(prev_group['rootsid']))
                if score > best_score:
                    best_match = prev_label
                    best_score = score
            if best_match and best_score > 0.3:  # Use a threshold to decide on a match
                current_labels[label] = best_match
                unmatched_labels.discard(label)

        # Assign new labels to unmatched labels
        for label in unmatched_labels:
            current_labels[label] = f"Region {new_region_start}"
            new_region_start += 1

        gdf[f'partition_{i}'] = gdf[f'partition_{i}'].map(current_labels)
    else:
        # Initialize partitions for the first map
        unique_partitions = gdf[f'partition_{i}'].unique()
        partition_to_region = {part: f"Region {j+1}" for j, part in enumerate(unique_partitions)}
        gdf[f'partition_{i}'] = gdf[f'partition_{i}'].map(partition_to_region)
    
    maps.append(gdf)

# Visualization of maps
fig, axes = plt.subplots(1, len(maps), figsize=(30, 10))

for i, gdf in enumerate(maps):
    gdf.plot(ax=axes[i], column=f'partition_{i}', legend=True)
    axes[i].set_title(f"Map {i + 1}")
    axes[i].set_axis_off()

plt.show()
```
